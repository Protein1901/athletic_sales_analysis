### 2. Determine which Region Sold the Most Products


#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".


# Show the top 5 results.
# Using groupby to calculate total products sold per region, state, and city
region_sales = (
    df_combined.groupby(['region', 'state', 'city'])['units_sold']
    .sum()
    .reset_index()
)
region_sales.rename(columns={'units_sold': 'total_products_sold'}, inplace=True)
region_sales.sort_values('total_products_sold', ascending=False, inplace=True)

# Display the top 5 regions with the most products sold
region_sales.head(5)
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.


# Rename the "units_sold" column to "Total_Products_Sold"


# Show the top 5 results.
# Create a pivot table to calculate total products sold for region, state, and city
region_state_city_pivot = (
    df_combined.pivot_table(
        values='units_sold',
        index=['region', 'state', 'city'],
        aggfunc='sum'
    )
    .reset_index()
    .rename(columns={'units_sold': 'Total_Products_Sold'})
    .sort_values(by='Total_Products_Sold', ascending=False)
)

# Display the top 5 results
region_state_city_pivot_top5 = region_state_city_pivot.head(5)
region_state_city_pivot_top5
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
# Filter the combined dataset for women's athletic footwear sales
womens_footwear_sales = df_combined[df_combined['product'] == "Women's Athletic Footwear"]

# Display the first few rows of the filtered dataset to verify
womens_footwear_sales.head()
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


# Show the top 5 results.
# Calculate the total number of women's athletic footwear sold grouped by retailer, region, state, and city
womens_footwear_units_sold = (
    womens_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold']
    .sum()
    .reset_index()
    .rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
    .sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
)

# Display the top 5 results
womens_footwear_units_sold_top5 = womens_footwear_units_sold.head(5)
womens_footwear_units_sold_top5
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.


# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"


# Show the top 5 results.
# Create a pivot table to calculate the total number of women's athletic footwear sold
womens_footwear_units_sold_pivot = (
    womens_footwear_sales.pivot_table(
        values='units_sold',
        index=['retailer', 'region', 'state', 'city'],
        aggfunc='sum'
    )
    .reset_index()
    .rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})
    .sort_values(by='Womens_Footwear_Units_Sold', ascending=False)
)

# Display the top 5 results
womens_footwear_units_sold_pivot_top5 = womens_footwear_units_sold_pivot.head(5)
womens_footwear_units_sold_pivot_top5
```

### 6. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.


# Optional: Rename the "total_sales" column to "Total Sales"


# Show the table.
# Create a pivot table to calculate total sales for women's athletic footwear by day
womens_footwear_daily_sales = (
    womens_footwear_sales.pivot_table(
        values='total_sales',
        index='invoice_date',
        aggfunc='sum'
    )
    .reset_index()
    .rename(columns={'total_sales': 'Total Sales'})
    .sort_values(by='Total Sales', ascending=False)
)

# Display the table sorted by total sales
womens_footwear_daily_sales.head()
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.


# Sort the resampled pivot table in descending order on "Total Sales".
# Resample the pivot table into daily bins and sum up the total sales
womens_footwear_daily_bins = (
    womens_footwear_sales.set_index('invoice_date')['total_sales']
    .resample('D')
    .sum()
    .reset_index()
    .rename(columns={'total_sales': 'Total Sales'})
    .sort_values(by='Total Sales', ascending=False)
)

# Display the top 5 days with the highest total sales
womens_footwear_daily_bins.head(5)
```

### 7.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.


# Sort the resampled pivot table in descending order on "Total Sales".
# Resample the pivot table into weekly bins and sum up the total sales
womens_footwear_weekly_bins = (
    womens_footwear_sales.set_index('invoice_date')['total_sales']
    .resample('W')
    .sum()
    .reset_index()
    .rename(columns={'total_sales': 'Total Sales'})
    .sort_values(by='Total Sales', ascending=False)
)

# Display the top 5 weeks with the highest total sales
womens_footwear_weekly_bins.head(5)
```
